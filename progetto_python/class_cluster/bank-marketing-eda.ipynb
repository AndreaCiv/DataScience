```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Bank Marketing.csv')
df.shape
```

```python
df.info()
```

```python
df.head()
```

# **Job and Subscription**

```python
plt.style.use('seaborn')
plt.barh(width=df['Job'].value_counts(),y=df['Job'].unique())
plt.title('Jobs')
plt.ylabel('Frequency')
plt.gca().invert_yaxis()
plt.show()
```

```python
df['Subscription'].replace(2,0,inplace=True)
subs = df['Subscription']
plt.style.use('seaborn')
plt.pie(subs.value_counts(),labels=subs.unique(), autopct='%.2f')
plt.title('Subscription Pie Chart')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(6,8))
data = df[['Job','Subscription']]
sns.countplot(data=data, y='Job', hue='Subscription')
plt.xscale('log')
plt.show()
```

# **Job and Education**

```python
plt.figure(figsize=(10, 10))
sns.countplot(data=df[['Job','Education']], y='Job', hue='Education')
plt.xscale('log')
plt.show()
```

# **Education and Subscription**

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(7, 7))
sns.countplot(data=df[['Subscription','Education']], y='Education', hue='Subscription')
plt.xscale('log')
plt.show()
```
